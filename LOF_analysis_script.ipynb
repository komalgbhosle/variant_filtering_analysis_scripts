```python
import pandas as pd 
```

```python
df = pd.read_csv('Info_column_splited_Asthma_variants.txt',sep = "\t")
```

```python
#Split INFO column by bar (|) and save it in k variable
k = df.INFO.str.split("|",expand=True,)
```

```python
#To count no of columns
len(k.axes[1]) 
```

```python
#To count no of rows
len(k.axes[0]) 
```

```python
#To add column names to splitted INFO columns
k.columns = ["Allele","Consequence","IMPACT","SYMBOL","Gene","Feature_type","Feature","BIOTYPE","EXON","INTRON","HGVSc","HGVSp","cDNA_position","CDS_position","Protein_position","Amino_acids","Codons","Existing_variation","DISTANCE","STRAND","FLAGS","VARIANT_CLASS","SYMBOL_SOURCE","HGNC_ID","CANONICAL","TSL","APPRIS","CCDS","ENSP","SWISSPROT","TREMBL","UNIPARC","GENE_PHENO","SIFT","PolyPhen","DOMAINS","HGVS_OFFSET","AF","AFR_AF","AMR_AF","EAS_AF","EUR_AF","SAS_AF","AA_AF","EA_AF","ExAC_AF","ExAC_Adj_AF","ExAC_AFR_AF","ExAC_AMR_AF","ExAC_EAS_AF","ExAC_FIN_AF","ExAC_NFE_AF","ExAC_OTH_AF","ExAC_SAS_AF","MAX_AF","MAX_AF_POPS","CLIN_SIG","SOMATIC","PHENO","PUBMED","MOTIF_NAME","MOTIF_POS","HIGH_INF_POS","MOTIF_SCORE_CHANGE"]
```

```python
k.head()
```

```python
# To filter only important columns from splitted INFO column and saved in p variable
p = k[["Consequence","SYMBOL","Gene","Feature_type","Feature","BIOTYPE","EXON","INTRON","Existing_variation","SIFT","PolyPhen"]]
```

```python
p.head()
```

```python
#join splitted INFO column with main vcf file 
final_INFO_splitted_file = df.join(p)
```

```python
final_INFO_splitted_file.head()
```

```python
# To remove any extra or unwanted columns from final_INFO_splitted_file
removed_columns = final_INFO_splitted_file.drop(['Unnamed: 7','INFO'],axis=1)
```

```python
removed_columns.head()
```

```python
#to check only missesnse variants
missense_variants = removed_columns[removed_columns['Consequence'].str.contains("missense_variant")] 
```

```python
missense_variants.head()
```

```python
#to count no of missense variants
len(missense_variants.axes[0]) 
```

```python
#To filter LOF mutations: missense varinats having "deleterious" effect in SIFT and "damaging" effect in polyphen database
LOF_mutations = removed_columns[removed_columns['Consequence'].str.contains("missense_variant") & removed_columns['SIFT'].str.contains("deleterious") & removed_columns['PolyPhen'].str.contains("damaging")]
```

```python
LOF_mutations.head()
```

```python
#To count no of LOF mutations
len(LOF_mutations.axes[0]) 
```

```python
# for rearranging columns of LOF output file 
def movecol(df, cols_to_move=[], ref_col='', place='After'):
    
    cols = df.columns.tolist()
    if place == 'After':
        seg1 = cols[:list(cols).index(ref_col) + 1]
        seg2 = cols_to_move
    if place == 'Before':
        seg1 = cols[:list(cols).index(ref_col)]
        seg2 = cols_to_move + [ref_col]
    
    seg1 = [i for i in seg1 if i not in seg2]
    seg3 = [i for i in cols if i not in seg1 + seg2]
    
    return(df[seg1 + seg2 + seg3])
Arranged_LOF_cols = movecol(LOF_mutations, 
             cols_to_move=["Consequence","SYMBOL","Gene","Feature_type","Feature","BIOTYPE","EXON","INTRON","Existing_variation","SIFT","PolyPhen"], 
             ref_col='FORMAT',
             place='After')
#cols_to_move : name of columns to move
#ref_col : column where we want to move columns
#place : columns to be moved after or Before ref_col
```

```python
Arranged_LOF_cols.head()
```

```python
#print output
Arranged_LOF_cols.to_csv('Asthma_LOF_mutations.csv')
```
