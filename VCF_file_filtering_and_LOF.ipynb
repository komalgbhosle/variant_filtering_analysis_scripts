```python
#join splitted INFO column by bar (|) with main vcf file 
final_INFO_splitted_file = df1.join(p)
```

```python
# for rearranging columns of LOF output file 
def movecol(df, cols_to_move=[], ref_col='', place='After'):
    
    cols = df.columns.tolist()
    if place == 'After':
        seg1 = cols[:list(cols).index(ref_col) + 1]
        seg2 = cols_to_move
    if place == 'Before':
        seg1 = cols[:list(cols).index(ref_col)]
        seg2 = cols_to_move + [ref_col]
    
    seg1 = [i for i in seg1 if i not in seg2]
    seg3 = [i for i in cols if i not in seg1 + seg2]
    
    return(df[seg1 + seg2 + seg3])
Arranged_cols = movecol(final_INFO_splitted_file, 
             cols_to_move=["Consequence","SYMBOL","Gene","Feature_type","Feature","BIOTYPE","EXON","INTRON","Existing_variation","SIFT","PolyPhen"], 
             ref_col='FORMAT',
             place='After')
```

```python
Arranged_cols.head()
```

```python
# To remove any extra or unwanted columns from Arranged_cols file
removed_columns = Arranged_cols.drop(['INFO'],axis=1)
```

```python
removed_columns.head()
```

```python
#print output file with all mutations and its consequences
removed_columns.to_csv('Asthma_final_mutations.csv')
```
